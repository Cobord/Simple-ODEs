```python
#adapted from https://wiseodd.github.io/techblog/2016/11/05/levelset-method/

import numpy as np
import matplotlib.pyplot as plt

class LevelSet():
    def __init__(self,grid_length,grid_width,phi,F,dt):
        self.grid_length=grid_length
        self.grid_width=grid_width
        self.F=F
        self.dt=dt
        self.phi=phi
    def update(self):
        dphi = np.gradient(self.phi)
        dphi_norm = np.sqrt(dphi[0]**2+dphi[1]**2)
        self.phi += self.dt * self.F * dphi_norm
        

grid_length=20
grid_width=20
phi = np.random.randn(grid_length, grid_width)
F = .2*np.ones((grid_length,grid_width))
dt = 1
my_level_set=LevelSet(grid_length,grid_width,phi,F,dt)

it=10
for i in range(it):
    plt.contour(phi, 0)
    plt.show()
    my_level_set.update()
```
